# Chebyshev center

Find the center and radius of the smallest sphere that encloses the polyhedron defined by the inequalities
$$ \left\{ 2 x - y + 2 z \le 2,\quad
-x + 2 y + 4 z \le 16,\quad
x + 2 y - 2 z \le 8,\quad
x \ge 0,\,
y \ge 0,\,
z \ge 0 \right\}$$

```julia
A = [2 -1 2; -1 2 4; 1 2 -2; -1 0 0; 0 -1 0; 0 0 -1];
b = [2; 16; 8; 0; 0; 0]

using JuMP, Clp

m = Model(solver=ClpSolver())
@variable(m, r >= 0)           # radius
@variable(m, x[1:3])           # coordinates of center
for i = 1:size(A,1)
    @constraint(m, A[i,:]'*x + r*norm(A[i,:]) <= b[i])
end
@objective(m, Max, r)     # maximize radius

status = solve(m)
center = getvalue(x)
radius = getvalue(r)

println(status)
println("The coordinates of the Chebyshev center are: ", center)
println("The largest possible radius is: ", radius)
```

```julia
using JuMP, PyPlot, Polyhedra, CDDLib

# plot the polyhedron
hrep = SimpleHRepresentation(A, b)
poly = polyhedron(hrep, CDDLibrary(:float))
points,faces,normals = Polyhedra.fulldecompose(poly);
x,y,z = [],[],[]
for p in points
    push!(x,p[1])
    push!(y,p[2])
    push!(z,p[3])
end
i,j,k = [],[],[]
for f in faces
    push!(i,f[1]-1)
    push!(j,f[2]-1)
    push!(k,f[3]-1)
end
```

```julia
t = mesh3d(x=x,y=y,z=z,i=i,j=j,k=k,opacity=0.6)


# plot the sphere
N = 32
kludge = 0.98    # shrink the radius slightly so it doesn't actually touch the polyhedron
u = linspace(0, 2π, N)
v = linspace(0, π, N)
x = center[1] + kludge * radius * cos(u) * sin(v)'
y = center[2] + kludge * radius * sin(u) * sin(v)'
z = center[3] + kludge * radius * repeat(cos(v)',outer=[N, 1])
s = surface(x=x,y=y,z=z,showscale=false)
;
```
